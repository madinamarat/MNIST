### Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
from pprint import pprint
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.vgg16 import decode_predictions
```

### Model

```python
m = VGG16(weights='imagenet')
```

## Using VGG16

### Data

```python
#lets pick an image
image = plt.imread('/Users/madinamukhambekova/Desktop/spearmint-vector-student-code/Week_09/img/aquarium.jpg')
plt.imshow(image)
```

### Preprocess - resize, and add a dimension

```python
image = cv2.resize(image,(224,224),interpolation=cv2.INTER_CUBIC)
image = np.expand_dims(image, axis=0)
image.shape
```

### Predict, and decode predictions

```python
results = m.predict(image)
```

```python
pprint(decode_predictions(results, 10))
```
